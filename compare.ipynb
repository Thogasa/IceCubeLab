```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('build/KNNRoc.txt', sep = ' ')
fprKnn = df['#fpr'].to_numpy()
tprKnn = df['tpr'].to_numpy()

df = pd.read_csv('build/BdtRoc.txt', sep = ' ')
fprbdt = df['#fpr'].to_numpy()
tprbdt = df['tpr'].to_numpy()

df = pd.read_csv('build/GNBRoc.txt', sep = ' ')
fprgnb = df['#fpr'].to_numpy()
tprgnb = df['tpr'].to_numpy()
```

```python
plt.plot(fprKnn[:-1], tprKnn[:-1], linewidth=2, label = 'kNN')
plt.plot(fprbdt[:-1], tprbdt[:-1], linewidth=2, label = 'BDT')
plt.plot(fprgnb[:-1], tprgnb[:-1], linewidth=2, label = 'Naive Bayes')

plt.plot([0,1], [0,1], 'k--' )
plt.rcParams['font.size'] = 12



props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)

plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.legend(loc = 'best')
plt.savefig('build/compareROC.pdf')
```

```python
df = pd.read_csv('build/KNNPRC.txt', sep = ' ')
trknn = df['thresh'].to_numpy()
fbknn = df['fb'].to_numpy()


df = pd.read_csv('build/BDTPRC.txt', sep = ' ')
trbdt = df['thresh'].to_numpy()
fbbdt = df['fb'].to_numpy()

df = pd.read_csv('build/GNBPRC.txt', sep = ' ')
trgnb = df['thresh'].to_numpy()
fbgnb = df['fb'].to_numpy()
```

```python
plt.plot(trknn, fbknn, linewidth=2, label = 'kNN, AUC = 0.967')
plt.plot(trbdt, fbbdt, linewidth=2, label = 'BDT, AUC = 0.994')
plt.plot(trgnb, fbgnb, linewidth=2, label = 'Naive Bayes, AUC = 0.926')

plt.rcParams['font.size'] = 12



props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)

plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.legend(loc = 'best')
plt.savefig('build/compareFB.pdf')
```
